Gray Level Slicing

```python
Gimg = cv2.imread('/content/img.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Define the intensity values for slicing
low = 100
high = 200

# Apply Gray Level Slicing
GLS_image = np.copy(Gimg)
GLS_image[(Gimg > low) & (Gimg < high)] = 255

# Display the original and processed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Gimg, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(GLS_image, cmap='gray')
plt.title('Gray Level Sliced Image')
plt.axis('off')

plt.show()
```

```python
# user defined
def gray_level_slicing(Gimg, low, high):
    GLSimg01 = np.copy(Gimg)
    GLSimg01[(Gimg > low) & (Gimg < high)] = 255
    return GLSimg01


# Define the intensity values for slicing
low = 100
high = 200

# Apply user-defined Gray Level Slicing function
GLSimg01 = gray_level_slicing(Gimg, low, high)

# Display the original and processed images using Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Gimg, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(GLSimg01, cmap='gray')
plt.title('Gray Level Sliced Image')
plt.axis('off')

plt.show()
```

Histograms Equalization

```python
#in built
equ = cv2.equalizeHist(Gimg)
equ
```

```python
# Display the original and equalized images using Matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(Gimg, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equ, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.show()
```

```python
#user defined
def histogram_equalization(image):
    hist, bins = np.histogram(image.flatten(), 256, [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max() / cdf.max()
    equalized_image = np.interp(image.flatten(), bins[:-1], cdf_normalized)
    equalized_image = equalized_image.reshape(image.shape).astype(np.uint8)
    return equalized_image


# Perform histogram equalization using the user-defined function
equ01 = histogram_equalization(Gimg)

# Display the original and equalized images using Matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(Gimg, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equ01, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.show()
```
