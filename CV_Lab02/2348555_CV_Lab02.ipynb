```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('/content/lab02.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
h, w = img_gray.shape

```

```python
# Identity transformation
mat_identity = np.float32([[1, 0, 0], [0, 1, 0]])
img1 = cv2.warpAffine(img_gray, mat_identity, (w, h))
cv2.imwrite('identity_transformation.jpg', img1)
```

```python
# Reflection transformation
mat_reflect = np.float32([[1, 0, 0], [0, -1, h]])
img2 = cv2.warpAffine(img_gray, mat_reflect, (w, h))
cv2.imwrite('reflection_transformation.jpg', img2)
```

```python
# Translation
tx, ty = 50, 50
mat_translate = np.float32([[1, 0, tx], [0, 1, ty]])
img3 = cv2.warpAffine(img_gray, mat_translate, (w, h))
cv2.imwrite('translation.jpg', img3)
```

```python
# Scaling
s_x, s_y = 0.75, 0.75
mat_scale = np.float32([[s_x, 0, 0], [0, s_y, 0]])
img4 = cv2.warpAffine(img_gray, mat_scale, (w, h))
cv2.imwrite('scaling.jpg', img4)
```

```python
# Rotation
theta = np.pi / 6
mat_rotate = cv2.getRotationMatrix2D((w / 2, h / 2), np.degrees(theta), 1)
img5 = cv2.warpAffine(img_gray, mat_rotate, (w, h))
cv2.imwrite('rotation.jpg', img5)
```

```python
# Shear
shear_value = 0.3
mat_shear = np.float32([[1, shear_value, 0], [shear_value, 1, 0]])
img6 = cv2.warpAffine(img_gray, mat_shear, (w, h))
cv2.imwrite('shear.jpg', img6)
```

```python
# Display images side by side
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs[0, 0].imshow(img1, cmap='gray')
axs[0, 0].set_title('Identity Transformation')
axs[0, 1].imshow(img2, cmap='gray')
axs[0, 1].set_title('Reflection Transformation')
axs[0, 2].imshow(img3, cmap='gray')
axs[0, 2].set_title('Translation')
axs[1, 0].imshow(img4, cmap='gray')
axs[1, 0].set_title('Scaling')
axs[1, 1].imshow(img5, cmap='gray')
axs[1, 1].set_title('Rotation')
axs[1, 2].imshow(img6, cmap='gray')
axs[1, 2].set_title('Shear')

for ax in axs.flat:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
