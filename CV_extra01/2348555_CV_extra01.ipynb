```python
#Using ImageIO
import imageio as iio

# read an image
img = iio.imread("/content/images.jpg")

# write it in a new format
iio.imwrite("g4g.png", img)
```

```python
#Using MatplotLib
# importing matplotlib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Read Images
img = mpimg.imread('/content/images.jpg')

# Output Images
plt.imshow(img)
```

```python
#Using PIL
# importing PIL
#Python Imaging Library
from PIL import Image

# Read image
img = Image.open('/content/images.jpg')

# Output Images
#img.show(img)
display(img)

# prints format of image
print(img.format)

# prints mode of image
# Eg- RGB, RFBA, CMYK, BGRetc.
print(img.mode)

# Prints the name of the file
print("Filename: ", img.filename)

# Prints the size as a width and height tuple (in pixels)
print("Size: ", img.size)

# Prints the width of the image (in pixels)
print("Width: ", img.width)

# Prints the height of the image (in pixels)
print("Height: ", img.height)

# Closing the image
img.close()

```
