```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
# Read image
img = cv2.imread('/content/img.jpg',0)
img01 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img01)
```

```python
# definig kernels
identity_kernel = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
edge_detection_kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
sharpening_kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
box_blur_kernel = np.ones((3, 3)) / 9
gaussian_blur_kernel = cv2.getGaussianKernel(5, 1)
```

image.png

Filter2D

```python
# identity kernel
dst_identity_kernel = cv2.filter2D(img.copy(), -1, identity_kernel)
# vis
plt.figure(figsize=(20, 10))

plt.subplot(3,2,1),plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(3,2,2),plt.imshow(dst_identity_kernel, cmap='gray'), plt.title('Identity (filter2D)')
```

```python
# edge detection kernel
dst_edge_detection_kernel = cv2.filter2D(img.copy(), -1, edge_detection_kernel)

# vis
plt.figure(figsize=(20, 10))

plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(3, 2, 2), plt.imshow(dst_edge_detection_kernel, cmap='gray'), plt.title('Edge detection (filter2D)')
```

```python
# sharpening kernel
dst_sharpening_kernel = cv2.filter2D(img.copy(), -1, sharpening_kernel)
#vis
plt.figure(figsize=(20, 10))
plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(3, 2, 2), plt.imshow(dst_sharpening_kernel, cmap='gray'), plt.title('Sharpen (filter2D)')
```

```python
# box blur kernel
dst_box_blur_kernel = cv2.filter2D(img.copy(), -1, box_blur_kernel)
#vis
plt.figure(figsize=(20, 10))
plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(3, 2, 2), plt.imshow(dst_box_blur_kernel, cmap='gray'), plt.title('Box blur (filter2D)')
```

```python
# Gaussian Blur Kernel
dst_gaussian_blur_kernel = cv2.filter2D(img.copy(), -1, gaussian_blur_kernel)
#vis
plt.figure(figsize=(20, 10))
plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(3, 2, 2), plt.imshow(dst_gaussian_blur_kernel, cmap='gray'), plt.title('Gaussian blur (filter2D)')
```

Convolve

```python
from scipy.ndimage import convolve
```

```python
# Define modes
modes = ['reflect', 'constant', 'nearest', 'mirror', 'wrap']
```

```python
# Convolve with identity kernel for each mode
for mode in modes:
    dst_identity_kernel = convolve(img.copy(), identity_kernel, mode=mode)
    plt.figure(figsize=(5, 5))
    plt.imshow(dst_identity_kernel, cmap='gray')
```
